```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
sheet=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
sheet.head(n=2)
```

```python
sheet.info()
```

```python
sheet['TotalCharges'].unique()
```

```python
def gender_category( gender):
    if gender=='Female':
        return 1
    else:
        return 0
sheet['Gen']=sheet['gender'].apply(gender_category)    
sheet.drop(axis=1,columns='gender',inplace=True)
def Partner_category(Partner):

    if Partner=='Yes':
        return 1
    else:
        return 0
sheet['partner']=sheet['Partner'].apply(Partner_category)    
sheet.drop(axis=1,columns='Partner',inplace=True)         
          

#convert Dependents to category
def Dependents_category(Dependents):
    if Dependents=='Yes':
        return 1
    else:
        return 0
sheet['Dep.']=sheet['Dependents'].apply(Dependents_category)    
sheet.drop(axis=1,columns='Dependents',inplace=True)




##   
def phone_ser_category(PhoneService):
    if PhoneService=='Yes':
        return 1
    else:
         return 0 
        
sheet['Phone']=sheet['PhoneService'].apply(phone_ser_category)    
sheet.drop(axis=1,columns='PhoneService',inplace=True)
sheet['Phone'].dropna(inplace=True)

##
def Multiline_category(MultipleLines):
    if MultipleLines =='Yes':
        return 1
    elif MultipleLines=='No':
        return 0
    else:
         return 0     
sheet['Multiphone']=sheet['MultipleLines'].apply(Multiline_category)    
sheet.drop(axis=1,columns='MultipleLines',inplace=True)
sheet['Multiphone'].dropna(inplace=True)


##
def Internet_category(InternetService):
    if InternetService =="DSL":   #DSL 
        return 1
    elif InternetService=="Fiber optic":  #Fiber optic
        return 0
    else:
         return 0     
sheet['Internet']=sheet['InternetService'].apply(Internet_category)    
sheet.drop(axis=1,columns='InternetService',inplace=True)
sheet['Internet'].dropna(inplace=True)
#
def O_security_category(OnlineSecurity):
    if OnlineSecurity=='Yes':
        return 1
    elif OnlineSecurity=='No':
        return 0
    else:
         return 0     
sheet['O.Security']=sheet['OnlineSecurity'].apply(O_security_category)    
sheet.drop(axis=1,columns='OnlineSecurity',inplace=True)
sheet['O.Security'].dropna(inplace=True)


#
def O_Backup_category(OnlineBackup):
    if OnlineBackup=='Yes':
        return 1
    elif OnlineBackup=='No':
        return 0
    else:
         return 0     
sheet['O.Backup']=sheet['OnlineBackup'].apply(O_Backup_category)    
sheet.drop(axis=1,columns='OnlineBackup',inplace=True)
sheet['O.Backup'].dropna(inplace=True)


#
def DevProtection_category(DeviceProtection):
    if DeviceProtection=='Yes':
        return 1
    elif DeviceProtection=='No':
        return 0
    else:
         return 0     
sheet['DevProtection']=sheet['DeviceProtection'].apply(DevProtection_category)    
sheet.drop(axis=1,columns='DeviceProtection',inplace=True)
sheet['DevProtection'].dropna(inplace=True)

##
def TechSupport_category(TechSupport):
    if TechSupport=='Yes':
        return 1
    elif TechSupport=='No':
        return 0
    else:
         return 0     
sheet['TechSup.']=sheet['TechSupport'].apply(TechSupport_category)    
sheet.drop(axis=1,columns='TechSupport',inplace=True)
sheet['TechSup.'].dropna(inplace=True)


##
def stream_tv_category(StreamingTV):
    if StreamingTV=='Yes':
        return 1
    elif StreamingTV=='No':
        return 0
    else:
         return 0     
sheet['Stream Tv.']=sheet['StreamingTV'].apply(stream_tv_category)    
sheet.drop(axis=1,columns='StreamingTV',inplace=True)
sheet['Stream Tv.'].dropna(inplace=True)



##
def stream_movies_category(StreamingMovies):
    if StreamingMovies =='Yes':
        return 1
    elif StreamingMovies=='No':
        return 0
    else:
         return 0    
sheet['Stream Movies.']=sheet['StreamingMovies'].apply(stream_movies_category)    
sheet.drop(axis=1,columns='StreamingMovies',inplace=True)
sheet['Stream Movies.'].dropna(inplace=True)



###
def Paper_less_Billing_category(PaperlessBilling):
    if PaperlessBilling =='Yes':
        return 1
    elif PaperlessBilling=='No':
        return 0
    else:
         return np.nan     
sheet['PaperBilling.']=sheet['PaperlessBilling'].apply(Paper_less_Billing_category)    
sheet.drop(axis=1,columns='PaperlessBilling',inplace=True) 
sheet['PaperBilling.'].dropna(inplace=True)


###
def Payment_Method_category(PaymentMethod):
    if PaymentMethod =='Bank transfer (automatic)':
        return 1.1
    elif  PaymentMethod =='Credit card (automatic)':
        return 1.2
    elif  PaymentMethod == 'Electronic check':
        return 2.1
    elif  PaymentMethod =='Mailed check':
        return 2.2
    else:
         return 0     
sheet['Payment']=sheet['PaymentMethod'].apply(Payment_Method_category)    
sheet.drop(axis=1,columns='PaymentMethod',inplace=True) 
sheet['Payment'].dropna(inplace=True)

###########
def Churn_category(Churn):
    if Churn=='Yes':
        return 1
    else:
         return 0 
        
        
sheet['Churn_1']=sheet['Churn'].apply(Churn_category)    
sheet.drop(axis=1,columns='Churn',inplace=True)
sheet['Churn_1'].dropna(inplace=True)

###########
def Co_category(Contract):
    if Contract=='Month-to-month':
        return 0
    elif Contract=='One year':
        return 1
    elif Contract=='Two year':
        return 2
    
    else:
         return 0
        
        
sheet['Con.']=sheet['Contract'].apply(Co_category)    
sheet.drop(axis=1,columns='Contract',inplace=True)
sheet['Con.'].dropna(inplace=True)



##########
```

```python
x=sheet.drop(['customerID','Phone','Multiphone','Churn_1','TotalCharges','Internet','MonthlyCharges','Payment'],axis='columns')
y=sheet.Internet
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
len(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor(n_neighbors=80)
model.fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)
y_hat
```

```python
mse=mean_squared_error(y_test,y_hat)*100

print("mse=",mse)
```
